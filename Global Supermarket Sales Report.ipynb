This notebook serves as a recreation of past projects showcasing a comprehensive business intelligence report developed for a Sales Dashboard. The report aims to provide key insights into sales, profit, and regional performance, supporting data-driven decision-making.

## Step-by-Step Report Creation Procedure

### 1. Data Acquisition and Understanding

The first step involved acquiring the 'Global Supermarket' dataset. This dataset typically includes transactional information such as sales figures, product categories, customer segments, order dates, shipping details, and geographical data (regions, countries). A thorough understanding of the dataset's structure, data types, and potential missing values was crucial at this stage.

Inspiration of design is from my previous work, patterned with the data from an open-source.

### 2. Data Cleaning and Transformation

Raw data often contains inconsistencies, errors, or requires reshaping to be suitable for analysis. This phase involved:

* **Handling Missing Values:** Addressing any nulls or incomplete records.
* **Data Type Correction:** Ensuring columns were in the correct format (e.g., dates as dates, numerical values as numbers).
* **Creating Calculated Columns:** Deriving new metrics essential for analysis, such as 'Profit %' or 'Shipping Days'
* **Standardizing Text Fields:** Ensuring consistency in categorical data (e.g., 'Consumer' vs. 'consumer').

### 3. Dashboard Design and Visualization

This is where the processed data was transformed into actionable insights through visual representation. The process involved:

* **Choosing Appropriate Visuals:** Selecting charts and graphs (e.g., line charts for trends, pie charts for proportions, bar charts for comparisons) that best convey the information.
* **Layout and Aesthetics:** Designing an intuitive and visually appealing layout for each dashboard tab.
* **Interactivity:** Implementing slicers, filters, and drill-through capabilities to allow users to explore data dynamically.
* **Iterative Refinement:** Testing and refining the dashboards based on usability and feedback to ensure clarity and impact.

### 4. Reporting and Deployment

The final step involved organizing the various dashboards into a cohesive report structure, making it ready for sharing and consumption. This Jupyter Notebook serves as an accessible way to showcase the report's design and key features.

## Report Dashboards

### 1. Main Dashboard

This dashboard provides a high-level overview of the global supermarket's performance, showcasing key metrics like total sales, profit, orders, and returns. It also visualizes sales trends over time and breaks down sales by customer segment and market.

### 2. Sales Dashboard

Focusing on sales performance, this dashboard provides detailed monthly trends, identifies top-selling products and countries, and highlights areas with the least sales, offering insights into sales drivers and areas for improvement.

### 3. Profit Analysis

This section is dedicated to understanding profitability. It details profit generated from different products and across various regions, crucial for identifying profitable ventures and areas requiring cost optimization.

### 4. Regional Operations Analytics

This dashboard offers a geographical perspective, breaking down sales, profit, and returns across major regions (Africa, Asia Pacific, Europe, LATAM, USCA). It also shows the distribution of product categories within each region, helping to understand regional market dynamics.

## Conclusion

This interactive report provides a robust framework for analyzing the Global Supermarket's operations. By leveraging data visualization, it empowers stakeholders to quickly grasp performance, identify trends, and make informed strategic decisions to optimize sales and profitability.
